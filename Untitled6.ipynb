```python
import numpy as np
import matplotlib.pyplot as plt

# 定数
V_1 = 200
r_2_prime = 3.92
x_total = 9.49  # x_1 + x_2_primeの合計
x_1 = x_2_prime = x_total / 2  # x_1とx_2_primeを等しく仮定
r_1 = 5.9506
Y_0 = 0.001948 + 1j * 0.0122
f = 50
p = 2
omega_s = 2 * np.pi * f / p

# すべりの範囲
s = np.linspace(0.01, 1, 100)

# 一次電流、力率、出力特性の計算
I_1_prime = []
I_1_total = []
cos_phi = []
P_1 = []
P = []
T = []
n = []
efficiency = []
I_0 = V_1 / np.sqrt(3) * Y_0

for slip in s:
    Z_2s = r_2_prime / slip + 1j * x_2_prime
    Z_1 = r_1 + 1j * x_1
    I1_prime = V_1 / np.sqrt(3) / (Z_1 + Z_2s)
    I_1_prime.append(np.abs(I1_prime))

    Y_total = Y_0 + 1 / (Z_1 + Z_2s)
    cos_phi.append(np.real(Y_total) / np.abs(Y_total))

    I1_total = V_1 / np.sqrt(3) * Y_total
    I_1_total.append(np.abs(I1_total))

    P1 = 3 * (V_1 / np.sqrt(3)) * np.abs(I1_total) * np.real(Y_total) / np.abs(Y_total)
    P_1.append(P1)

    P1c = 3 * np.abs(I1_prime)**2 * r_1
    P2c = 3 * np.abs(I1_prime)**2 * r_2_prime

    P2 = 3 * np.abs(I1_prime)**2 * r_2_prime / slip
    P_out = P2 * (1 - slip)

    P.append(P_out)
    T.append(P_out / omega_s)
    n.append((1 - slip) * 1500)

    eff = P_out / P1 * 100
    efficiency.append(eff)

# 速度特性のプロット（すべりを左右反転）
fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.plot(1 - s, T, 'b-', label='Torque [Nm]')
ax1.set_xlabel('Slip')
ax1.set_ylabel('Torque [Nm]', color='b')
ax1.tick_params('y', colors='b')

ax2 = ax1.twinx()
ax2.plot(1 - s, I_1_total, 'r-', label='Current [A]')
ax2.set_ylabel('Current [A]', color='r')
ax2.tick_params('y', colors='r')

fig.tight_layout()
fig.suptitle('Torque and Current vs Slip (Reversed)', y=1.02)
plt.show()
# 出力特性のプロット
fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(P, n, 'b-', label='Speed [rpm]')
ax1.set_xlabel('Output Power [W]')
ax1.set_ylabel('Speed [rpm]', color='b')
ax1.tick_params('y', colors='b')

ax2 = ax1.twinx()
ax2.plot(P, I_1_total, 'r-', label='Current [A]')
ax2.set_ylabel('Current [A]', color='r')
ax2.tick_params('y', colors='r')

ax3 = ax1.twinx()
ax3.spines['right'].set_position(('outward', 60))
ax3.plot(P, T, 'g-', label='Torque [Nm]')
ax3.set_ylabel('Torque [Nm]', color='g')
ax3.tick_params('y', colors='g')

ax4 = ax1.twinx()
ax4.spines['right'].set_position(('outward', 120))
ax4.plot(P, efficiency, 'c-', label='Efficiency [%]')
ax4.set_ylabel('Efficiency [%]', color='c')
ax4.tick_params('y', colors='c')

ax5 = ax1.twinx()
ax5.spines['right'].set_position(('outward', 180))
ax5.plot(P, cos_phi, 'm-', label='Power Factor')
ax5.set_ylabel('Power Factor', color='m')
ax5.tick_params('y', colors='m')

fig.tight_layout()
fig.suptitle('Output Characteristics', y=1.02)
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))
plt.show()
```
